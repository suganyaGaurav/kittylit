```python
import sqlite3

# Connect to (or create) the SQLite database file in your project data folder
conn = sqlite3.connect('data/kittylit_books.db')  # relative path from your project root

cursor = conn.cursor()

# Create table if it doesn't exist
cursor.execute('''
CREATE TABLE IF NOT EXISTS books (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT NOT NULL,
    author TEXT,
    year INTEGER,
    description TEXT,
    isbn TEXT UNIQUE,
    thumbnail_url TEXT,
    source TEXT,
    added_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
)
''')

conn.commit()
conn.close()

print("Database and table created successfully!")
```

```python
import sqlite3

# Connect to the existing database
conn = sqlite3.connect('data/kittylit_books.db')
cursor = conn.cursor()

# Define a sample book record as a tuple
sample_book = (
    'The Power of AI',            # title
    'Ammu Engineer',              # author
    2025,                        # year
    'An inspiring book on AI projects and engineering',  # description
    '9781234567897',             # isbn (unique)
    'http://example.com/image.jpg',  # thumbnail_url
    'live_api'                   # source (where the data comes from)
)

# Insert the sample record into the books table
cursor.execute('''
INSERT OR IGNORE INTO books (title, author, year, description, isbn, thumbnail_url, source)
VALUES (?, ?, ?, ?, ?, ?, ?)
''', sample_book)

# Commit the transaction to save changes
conn.commit()

# Close the connection
conn.close()

print("Sample book inserted successfully!")
```

```python
import sqlite3

# Connect to the database
conn = sqlite3.connect('data/kittylit_books.db')
cursor = conn.cursor()

# Execute a query to fetch all records from books table
cursor.execute('SELECT * FROM books')

# Fetch all rows from the executed query
rows = cursor.fetchall()

# Loop through rows and print each book record
for row in rows:
    print(f"ID: {row[0]}")
    print(f"Title: {row[1]}")
    print(f"Author: {row[2]}")
    print(f"Year: {row[3]}")
    print(f"Description: {row[4]}")
    print(f"ISBN: {row[5]}")
    print(f"Thumbnail URL: {row[6]}")
    print(f"Source: {row[7]}")
    print(f"Added At: {row[8]}")
    print("-" * 40)

# Close the connection
conn.close()
```
